## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data.head()
```

```python
# Count of unique items purchased
item_count = len(purchase_data["Item Name"].unique())

# Total revenue and formatted revenue to add $
revenue = sum(purchase_data["Price"])
formatted_revenue = '${:,.2f}'.format(revenue)

# Total number of purchases, along with average price. Formatted average price to add $
purchase_count = len(purchase_data["Price"])
avg_price = revenue/purchase_count
formatted_avg_price = '${:,.2f}'.format(avg_price)
formatted_avg_price
```

```python
# Purchase Summary Dataframe
purchase_summary_df = pd.DataFrame({"Number of Unique Itmes": [item_count],
                              "Average Price": formatted_avg_price,
                              "Number of Purchases": purchase_count,
                              "Total Revenue": formatted_revenue})
purchase_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_data.head()
```

```python
# Created Dataframe of Male Users, assigned variable with count of male users
male_purchase_data = purchase_data.loc[purchase_data.loc[:, 'Gender'] == 'Male',:]
male_count = male_purchase_data['SN'].nunique()

# Percentage of male users
male_percent = male_count/player_count
formatted_male_percentage = '{:,.2f}%'.format(male_percent * 100)

# Created Dataframe of Female Users, assigned variable with count of female users
female_purchase_data = purchase_data.loc[purchase_data.loc[:, 'Gender'] == 'Female',:]
female_count = female_purchase_data['SN'].nunique()

# Percentage of female users
female_percent = female_count/player_count
formatted_female_percentage = '{:,.2f}%'.format(female_percent * 100)

# Created Dataframe of Other/Non-binary Users, assigned variable with count of other/non-binary users
other_purchase_data = purchase_data.loc[purchase_data.loc[:, 'Gender'] == 'Other / Non-Disclosed',:]
other_count = other_purchase_data['SN'].nunique()

# Percentage of other users
other_percent = other_count/player_count
formatted_other_percentage = '{:,.2f}%'.format(other_percent * 100)
```

```python
# Dataframe created to summarize users by gender

gender_summary_df = pd.DataFrame(
    {"Gender": ["Male", "Female", "Other/Non-Disclosed"],
     "Total Count": [male_count, female_count, other_count],
     "Percentage of Players": [formatted_male_percentage, formatted_female_percentage, formatted_other_percentage]})
gender_summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data.head()
```

```python
#Defined age group counts for overall and unique, then pulled in the purchase count and purchase amounts to do division on.

age_groups = purchase_data.groupby("Age Group")
age_groups_unique = age_groups['SN'].nunique()
purchase_count_by_age = age_groups["Purchase ID"].count()
total_purchases_by_age = age_groups['Price'].sum()
average_price_by_age = total_purchases_by_age/purchase_count_by_age
average_price_per_person = total_purchases_by_age/age_groups_unique
purchase_count_by_age
```

```python
# Created summary table dataframe and vectorized the functions to create proper formatting.
age_purchase_summary_df = pd.DataFrame({"Purchase Count":purchase_count_by_age,
                                        "Average Purchase Price": average_price_by_age,
                                        "Total Purchase Value":total_purchases_by_age,
                                        "Average Total Purchase Per Person":average_price_per_person
                                        }
)

age_purchase_summary_df["Average Purchase Price"] = age_purchase_summary_df["Average Purchase Price"].map('${:,.2f}'.format)
age_purchase_summary_df["Total Purchase Value"] = age_purchase_summary_df["Total Purchase Value"].map('${:,.2f}'.format)
age_purchase_summary_df["Average Total Purchase Per Person"] = age_purchase_summary_df["Average Total Purchase Per Person"].map('${:,.2f}'.format)
age_purchase_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
# Created a dataframe grouped by the count of purchases

unique_purchase_count_df = pd.DataFrame(purchase_data.groupby(['SN']).count()['Purchase ID'])
unique_purchase_count_df.rename(columns={"Purchase ID":"Purchase Count"}, inplace=True)

# Created a second dataframe with the sum of prices
user_purchase_total = pd.DataFrame(purchase_data.groupby(['SN']).sum()['Price'])
unique_purchase_count_df
```

```python
# Merged the two above dataframes on the username
merged_df = pd.merge(unique_purchase_count_df,user_purchase_total,on="SN", how="outer")

# Renamed the price column to reflect the total purchase value
merged_df.rename(columns={"Price":"Total Purchase Value"}, inplace=True)

# Calculated the average price
merged_df["Average Purchase Price"] = merged_df["Total Purchase Value"]/merged_df["Purchase Count"]

# Sorted values by purchase value
purchase_count_sort = merged_df.sort_values("Total Purchase Value",ascending=False)

# Formatted floats to currency
purchase_count_sort["Average Purchase Price"] = purchase_count_sort["Average Purchase Price"].map('${:,.2f}'.format)
purchase_count_sort["Total Purchase Value"] = purchase_count_sort["Total Purchase Value"].map('${:,.2f}'.format)

purchase_count_sort.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
# Created dataframe to hold the items and the number of purchases
popular_item_df = pd.DataFrame(purchase_data.groupby(['Item ID', 'Item Name']).count()['Purchase ID'])

# Renamed column to reflect the purchase count
popular_item_df.rename(columns={"Purchase ID":"Purchase Count"}, inplace=True)

# Created a dataframe to hold the average price of each item
average_item_df = pd.DataFrame(purchase_data.groupby(['Item ID', 'Item Name']).mean()['Price'])
average_item_df.head()
```

```python
# Created a dataframe to merge the two previous dataframes into one.
most_popular_items_df = pd.merge(popular_item_df, average_item_df, left_index=True, right_index=True)

# New column created to calculate the total purchase value
most_popular_items_df['Total Purchase Value'] = most_popular_items_df['Purchase Count'] * most_popular_items_df['Price']
most_popular_items_df_sorted = most_popular_items_df.sort_values("Purchase Count",ascending=False)

# Formatted to change from float to currency
most_popular_items_df_sorted['Total Purchase Value'] = most_popular_items_df_sorted['Total Purchase Value'].map('${:,.2f}'.format)
most_popular_items_df_sorted['Price'] = most_popular_items_df_sorted['Price'].map('${:,.2f}'.format)
most_popular_items_df_sorted.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting

```python
# Took last main dataframe and sorted by total purchase value
most_profitable_items = most_popular_items_df.sort_values('Total Purchase Value',ascending=False)

# Formatted to change from float to currency
most_profitable_items['Total Purchase Value'] = most_profitable_items['Total Purchase Value'].map('${:,.2f}'.format)
most_profitable_items['Price'] = most_profitable_items['Price'].map('${:,.2f}'.format)

most_profitable_items.head()
```
